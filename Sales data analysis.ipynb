```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('sales_data.csv')
print(df)
```

```python
#Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
# Bar Chart: Revenue by Product
plt.figure(figsize=(10,5))
sns.barplot(x='Product', y='Revenue',data=df,estimator=sum)
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.title("Revenue by Product")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram: Distribution of Customer_Age
plt.figure(figsize=(8,5))
plt.hist(df['Customer_Age'], bins=20, edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Distribution of Customer Age")
plt.grid()
plt.show()
```

```python
# Box Plot: Profit Distribution Across Cities
plt.figure(figsize=(10,5))
sns.boxplot(x='City', y='Profit', data=df)
plt.xlabel("City")
plt.ylabel("Profit")
plt.title("Profit Distribution Across Cities")
plt.xticks(rotation=45)
plt.show()
```
